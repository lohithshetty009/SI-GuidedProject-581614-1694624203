Task1-Create a pandas dataframe (DataFrame name as 'df') with numpy random values (4 features and 4 observation)


```python
import pandas as pd
import numpy as np
import random

alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

data = {
    'Goals': np.random.randint(0, 100, size=4),
    'Assists': np.random.randint(0, 50, size=4),
    'Shots on Target': np.random.randint(10, 100, size=4),
    'Pass Completion': np.random.uniform(50.0, 95.0, size=4),
    'Club': [random.choice(['Club A', 'Club B', 'Club C']) for _ in range(4)]
}

df = pd.DataFrame(data)
print(df)
```

Rename the task - 1 'df' dataframe column names to 'Random value 1', 'Random value 2', 'Random value 3' & 'Random value 4

```python
df.rename(columns={'Goals':'Random value 1','Assists':'Random value 2','Shots on Target':'Random value 3','Pass Completion':'Random value 4','Club':'Random value 5'},inplace=True)
df
```
